```python
import math
import collections

import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
monalisa = np.load('monalisa.npy')
```

```python
monalisa.shape
```

```python
pp.imshow(monalisa)
```

```python
monalisa[600,400,0] #0 for the red component
```

```python
pp.imshow(monalisa[100:500,200:600,0:3])
```

```python
pp.imshow(monalisa[100:500,:,:]) # grab the entire range of a certain axis
```

```python
pp.imshow(monalisa[100:500,...]) #shorthand
```

```python
pp.imshow(monalisa[::20,::20,:]) #step -> take every 20th pixel -> reduce resolution
```

```python
pp.imshow(monalisa[::-20,::20,:])
```

```python
monalisa[20:300,20:300,:] = 255
```

```python
pp.imshow(monalisa)
```

```python
monalisa[20:300,20:300,:] = np.random.randint(100,255,size=(280,280,3))
```

```python
pp.imshow(monalisa)
```

### python list vs numpy array:

python list: slice a list -> make a copy of it  
numpy array: slice an array -> a new array pointing to same area of memory
